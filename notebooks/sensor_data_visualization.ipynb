## Time series data plotting

```python
from datetime import datetime
```

```python
datetime(2012, 5, 3, 19, 00, 5)
```

```python
def onlyDigits(st):
    retList = []
    if ord(st[0]) > 48 and ord(st[0]) <= 57:
        retList.append(st[0])
    
    for i in range(1, len(st)):
        if ord(st[i]) >= 48 and ord(st[i]) <= 57:
            retList.append(st[i])
            
    return ''.join(retList)
```

#### Producing Time Series data

```python
time_series  = []
for i in df['sess_time']:
    year = int(0 if onlyDigits(i[0:4].lstrip().rstrip()) == "" else onlyDigits(i[0:4].lstrip().rstrip()))
    month = int(0 if onlyDigits(i[5:7].lstrip().rstrip()) == "" else onlyDigits(i[5:7].lstrip().rstrip()))
    day = int(0 if onlyDigits(i[8:10].lstrip().rstrip()) == "" else onlyDigits(i[8:10].lstrip().rstrip()))
    hours = int(0 if onlyDigits(i[11:13].lstrip().rstrip()) == "" else onlyDigits(i[11:13].lstrip().rstrip()))
    minut = int(0 if onlyDigits(i[14:16].lstrip().rstrip()) == "" else onlyDigits(i[14:16].lstrip().rstrip()))
    seconds = int(0 if onlyDigits(i[17:19].lstrip().rstrip()) == "" else onlyDigits(i[17:19].lstrip().rstrip()))    
    #print(hours)
    tmp = datetime(year, month, day, hours, minut, seconds)
    time_series.append(tmp)
```

```python
len(time_series)
```

```python
num_records = 100
```

```python
x = np.array(time_series)[1:num_records]
sound = np.array(df['sound'])[1:num_records]
cum_dist = np.array(df['cum_dist'])[1:num_records]
temperature = np.array(df['temperature'])[1:num_records]
```

```python
x
```

```python
plt.xlabel("Time Series")
plt.ylabel("Variables")
plt.gca().set_color_cycle(['red', 'green', 'blue'])
plt.plot(x, sound, marker=r'$\clubsuit$')
plt.plot(x, cum_dist, marker=r'$\clubsuit$')
plt.plot(x, temperature, marker=r'$\clubsuit$')
plt.legend(['y = sound', 'y = distance ', 'y = temperature '], loc='upper left')
# plt.yticks(np.arange(min(y), 5,  0.5))
plt.show()
```
